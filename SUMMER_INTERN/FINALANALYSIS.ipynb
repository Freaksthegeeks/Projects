# **HANDLING OUTLIERS**

---



**FINDING THE OUTLIERS(EXTREME VALUES) OF THE FIELD(DAYS WITH AQI) IN A DATA SET:**

```python
from scipy import stats

column_of_interest = 'Days with AQI'
z_scores = stats.zscore(df[column_of_interest])
threshold = 3

# Use boolean indexing directly on the Series
df_no_outliers = df[(abs(z_scores) < threshold)]

# Update the original dataframe
df.update(df_no_outliers)

outliers = df[(abs(z_scores) >= threshold)]
non_outliers = df[abs(z_scores) < threshold]
df = non_outliers.copy()

print(outliers)

```
